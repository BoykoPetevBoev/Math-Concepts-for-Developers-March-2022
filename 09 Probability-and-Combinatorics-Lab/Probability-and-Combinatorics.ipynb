```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

## Probability and Combinatorics

```python
get_numbers = np.random.randint(low=1, high=10, size=1000)
```

```python
get_numbers
```

```python
plt.plot(range(1, len(get_numbers) + 1), get_numbers)
plt.xlabel("Range")
plt.ylabel('Number')
plt.show()
```

```python
get_numbers == 3
```

```python
data = [(num, (get_numbers == num).sum() / len(get_numbers)) for num in range(1, 10)]
```

```python
plt.hist(get_numbers, bins=range(1, 10))
```

```python
plt.bar([x[0] for x in data], [x[1] for x in data])
plt.show()
```
