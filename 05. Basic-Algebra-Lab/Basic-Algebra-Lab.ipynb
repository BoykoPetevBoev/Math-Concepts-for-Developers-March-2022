```python
import numpy.polynomial.polynomial as p
import numpy as np
import sympy
import matplotlib.pyplot as plt
```

**np.linspace** is a NumPy function that generates evenly spaced numbers between a start and an end value, inclusive. 
It takes three arguments:
- **start**: The starting value of the sequence.
- **stop**: The ending value of the sequence.
- **num (optional, default=50)**: The number of evenly spaced values to generate between 

**sympy.symbols** is a function in the SymPy library that creates one or more symbolic variables. It takes one or more string arguments, where each argument represents the name of a variable.

## Polynomials

```python
p.polyadd([-8, 5, 2], [-2, 0, 0, 0, 3])
```

```python
p.polymul([-8, 5, 2], [-2, 0, 0, 0, 3])
```

```python
polynomial = p.Polynomial([-2, 3, 5, 1, 3])
sympy.init_printing()
poly = reversed(polynomial.coef)
x = sympy.symbols('x')
sympy.Poly(poly, x).as_expr()
```

## Sets


```python
positive_x = {x for x in range(-5, 5) if x >= 0}
print(positive_x)
```

```python
set1 = { 1, 2, 3, 4 }
set2 = {3, 4, 5, 10, 3, 5, 10, 3, 3}

print(len(set2)) # 4
print(1 in set1) # True
print(10 not in set1) # True
print({1, 2}.issubset(set1)) # True
print(set1.union(set2)) # {1, 2, 3, 4, 5, 10}
print(set1.difference(set2)) # {1, 2}
print(set2.difference(set1)) # {10, 5}
print(set1.symmetric_difference(set2)) # {1, 2, 5, 10}
```

## Functions

```python

def plot_function(f, x_min = -10, x_max = 10, n_values = 2000):
  x = np.linspace(x_min, x_max, n_values)
  y = f(x)
  plt.plot(x, y)
  plt.show()
```

```python
plot_function(lambda x: np.sin(x))
```

```python
plot_function(lambda x: np.tan(x))
```

```python
def plot_function(f, x_min = -10, x_max = 10, n_values = 1000):
  plt.gca().set_aspect("equal")
  x = np.linspace(x_min, x_max, n_values)
  y = f(x)
  plt.plot(x, y)

plot_function(lambda x: np.sqrt(1 - x**2), -1, 1)
plot_function(lambda x: -np.sqrt(1 - x**2), -1, 1)
plt.show()
```
